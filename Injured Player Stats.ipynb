```python
def get_playerdf(url):
    r = requests.get(url)
    html_doc = r.text
    bsoup = BeautifulSoup(html_doc)
        
    bsoup.findAll('tr', limit=2)
    
    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in bsoup.findAll('tr', limit=2)[0].findAll('th')]
   
    # exclude the first column as we will not need the ranking order from Basketball Reference for the analysis
    headers = headers[1:]
    
    # avoid the first header row
    rows = bsoup.findAll('tr')[1:]
    player_stats = [[td.getText() for td in rows[i].findAll('td')]
            for i in range(len(rows))]
    
    bsoup_df = pd.DataFrame(player_stats, columns = headers)
    
    return bsoup_df
```

```python
def get_player_name(url):
    r = requests.get(url)
    html_doc = r.text
    bsoup = BeautifulSoup(html_doc)
           
    # use getText()to extract the text we need into a list
    name = [header.getText() for header in bsoup.findAll('h1', limit=2)]
    
    return name
```
