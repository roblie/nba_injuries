```python
import pandas as pd
import matplotlib as plt
```

```python
filename = 'nba-injuries-2010-2018/injuries.csv'
```

```python
injuries_df = pd.read_csv(filename, parse_dates=['Date'])
injuries_df.head()
```

```python
injuries_df.info()
```

```python
injuries_df = injuries_df.set_index('Date')
```

```python
injuries_df = injuries_df[injuries_df['Relinquised'].notnull()]
injuries_df
```

```python
injuries_df = injuries_df[['Team', 'Relinquised', 'Notes']]
injuries_df
```

```python
injuries_df.info()
```

```python
injuries_df[injuries_df['Team'].isnull()]
```

```python
injuries_df = injuries_df[injuries_df['Team'].notnull()]
injuries_df
```

```python
injuries_df.info()
```

```python
group_by = injuries_df.groupby(['Notes'])
count_group = group_by['Relinquised'].count()
print(count_group.sort_values(ascending=False))
```

```python
len(count_group)
```

```python
%matplotlib notebook
bar_graph = count_group[:20].plot(kind='bar')
```
