```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def get_soup(link):
    
    r = requests.get(link)
    html_doc = r.text
    bsoup = BeautifulSoup(html_doc)

    return bsoup
```

```python
def games_header(bsoup):

    bsoup.findAll('tr', limit=2)

    # use getText()to extract the text we need into a list
    headers = [th.getText() for th in bsoup.findAll('tr', limit=2)[0].findAll('th')]
        
    return headers
```

```python
def nba_games(bsoup):
    # avoid the first header row
    rows = bsoup.findAll('tr')[1:]
    game_info = [[td.getText() for td in rows[i].findAll('td')] for i in range(len(rows))]
    dates = [[th.getText() for th in bsoup.findAll('tr')[i].findAll('th')] for i in range(1,len(rows)+1)] 
    
    # combine dates and game_info
    games = [dates[i] + game_info[i] for i in range(len(rows))]
           
    return games
```

```python
nba_header = []
nba_games_2017 = []
year = '2017'
month = ['october', 'november', 'december', 'january', 'february', 'march', 'april', 'may', 'june']
url = ['https://www.basketball-reference.com/leagues/NBA_' + year + '_games-{}.html'.format(item) for item in month]
for link in url:
    nba_soup = get_soup(link)
    nba_header = games_header(nba_soup)
    nba_games_2017 += nba_games(nba_soup)
```

```python
nba_games_2018 = []
year = '2018'
month = ['october', 'november', 'december', 'january', 'february', 'march', 'april', 'may', 'june']
url = ['https://www.basketball-reference.com/leagues/NBA_' + year + '_games-{}.html'.format(item) for item in month]
for link in url:
    nba_soup = get_soup(link)
    nba_games_2018 += nba_games(nba_soup)
```
